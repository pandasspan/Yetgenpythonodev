# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
# 1.Problemin Çözümünü Buraya Yazınız
z = int(input("kaç adet ürün gireceğinizi yazınız"))
program = { }
for i in range( 1 , z+1):  
    x = input("ürün isim")
    y = float(input("ürün fiyat"))
    program.update({x : y})


liste = program.items()
with open("urunler.txt", "r+", encoding = "utf-8") as file:
# context manager
    yazılcak = str(liste)
    file.write(yazılcak)
with open("urunler.txt", "r+", encoding = "utf-8") as file:
    print(file.readlines())
````
